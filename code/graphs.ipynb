```python
# importing packages
import pandas as pd
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
```

```python
data = X.copy()
data.rename(
    columns={
        'age': 'Age', 
        'sex': 'Sex', 
        'cp': 'Chest Pain', 
        'trestbps': 'Resting BP', 
        'chol': 'Cholesterol', 
        'fbs': 'Fasting Blood Sugar', 
        'restecg': 'Resting ECG', 
        'thalach': 'Maximum HR', 
        'exang': 'Exercise Induced Angina', 
        'oldpeak': 'ST Depression', 
        'slope': 'ST/HR Slope', 
        'ca': 'Flourosopy Colored Vessels', 
        'thal': 'Thalassemia'
    }, inplace=True
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
data = X.copy()
data.rename(
    columns={
        'age': 'Age', 
        'sex': 'Sex', 
        'cp': 'Chest Pain Type', 
        'trestbps': 'Resting Blood Pressure', 
        'chol': 'Serum Cholesterol', 
        'fbs': 'Fasting Blood Sugar', 
        'restecg': 'Resting Electrocardiographic Results', 
        'thalach': 'Maximum Heart Rate Achieved', 
        'exang': 'Presence of Exercise Induced Angina', 
        'oldpeak': 'ST Depression Induced by Exercise on an ECG', 
        'slope': 'ST/HR Slope of the Peak Exercise ST Segment', 
        'ca': 'Number of Major Vessels Colored by Flourosopy', 
        'thal': 'Thalassemia'
    }, inplace=True
)

# Convert columns to categorical
data['Sex'] = data['Sex'].astype('category')
data['Chest Pain Type'] = data['Chest Pain Type'].astype('category')
data['Fasting Blood Sugar'] = data['Fasting Blood Sugar'].astype('category')
data['Resting Electrocardiographic Results'] = data['Resting Electrocardiographic Results'].astype('category')
data['Presence of Exercise Induced Angina'] = data['Presence of Exercise Induced Angina'].astype('category')
data['ST/HR Slope of the Peak Exercise ST Segment'] = data['ST/HR Slope of the Peak Exercise ST Segment'].astype('category')
data['Number of Major Vessels Colored by Flourosopy'] = data['Number of Major Vessels Colored by Flourosopy'].astype('category')
data['Thalassemia'] = data['Thalassemia'].astype('category')

# Rename categorical values
data['Sex'] = data['Sex'].cat.rename_categories({1: 'Male', 0: 'Female'})
data['Chest Pain Type'] = data['Chest Pain Type'].cat.rename_categories({1: 'Typical Angina', 2: 'Atypical Angina', 3: 'Non-Anginal Pain', 4: 'Asymptomatic'})
data['Fasting Blood Sugar'] = data['Fasting Blood Sugar'].cat.rename_categories({1: '>120mg/ml', 0: '<120mg/ml'})   
data['Resting Electrocardiographic Results'] = data['Resting Electrocardiographic Results'].cat.rename_categories({0: 'Normal', 1: 'ST-T wave abnormality', 2: 'Left Ventricular Hypertrophy'})
data['Presence of Exercise Induced Angina'] = data['Presence of Exercise Induced Angina'].cat.rename_categories({1: 'Yes', 0: 'No'})
data['ST/HR Slope of the Peak Exercise ST Segment'] = data['ST/HR Slope of the Peak Exercise ST Segment'].cat.rename_categories({1: 'Upsloping', 2: 'Flat', 3: 'Downsloping'})
data['Number of Major Vessels Colored by Flourosopy'] = data['Number of Major Vessels Colored by Flourosopy'].cat.rename_categories({0: '0', 1: '1', 2: '2', 3: '3', 4: '4'})
data['Thalassemia'] = data['Thalassemia'].cat.rename_categories({3: 'Normal', 6: 'Fixed Defect', 7: 'Reversible Defect'})
```

```python
# Define which columns are categorical and which are numerical
categorical_columns = data.select_dtypes(include=['category']).columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Plot categorical columns
for col in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=data[col])
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# Plot numerical columns
for col in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(data[col], kde=True)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Plot boxplots for numerical columns
for col in numerical_columns:
    plt.figure(figsize=(10, 1.5))
    sns.boxplot(x=data[col])
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.show()
```
