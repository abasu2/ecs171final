```python
# importing packages
import pandas as pd
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets
```

```python
data_with_target = X.copy()
data_with_target['Heart Disease'] = y
data_with_target.rename(
    columns={
        'age': 'Age', 
        'sex': 'Sex', 
        'cp': 'Chest Pain Type', 
        'trestbps': 'Resting Blood Pressure', 
        'chol': 'Serum Cholesterol', 
        'fbs': 'Fasting Blood Sugar', 
        'restecg': 'Resting Electrocardiographic Results', 
        'thalach': 'Maximum Heart Rate Achieved', 
        'exang': 'Presence of Exercise Induced Angina', 
        'oldpeak': 'ST Depression Induced by Exercise on an ECG', 
        'slope': 'ST/HR Slope of the Peak Exercise ST Segment', 
        'ca': 'Number of Major Vessels Colored by Flourosopy', 
        'thal': 'Thalassemia'
    }, inplace=True
)

# Convert columns to categorical
data_with_target['Sex'] = data_with_target['Sex'].astype('category')
data_with_target['Chest Pain Type'] = data_with_target['Chest Pain Type'].astype('category')
data_with_target['Fasting Blood Sugar'] = data_with_target['Fasting Blood Sugar'].astype('category')
data_with_target['Resting Electrocardiographic Results'] = data_with_target['Resting Electrocardiographic Results'].astype('category')
data_with_target['Presence of Exercise Induced Angina'] = data_with_target['Presence of Exercise Induced Angina'].astype('category')
data_with_target['ST/HR Slope of the Peak Exercise ST Segment'] = data_with_target['ST/HR Slope of the Peak Exercise ST Segment'].astype('category')
data_with_target['Number of Major Vessels Colored by Flourosopy'] = data_with_target['Number of Major Vessels Colored by Flourosopy'].astype('category')
data_with_target['Thalassemia'] = data_with_target['Thalassemia'].astype('category')
data_with_target['Heart Disease'] = data_with_target['Heart Disease'].astype('category')

# Rename categorical values
data_with_target['Sex'] = data_with_target['Sex'].cat.rename_categories({1: 'Male', 0: 'Female'})
data_with_target['Chest Pain Type'] = data_with_target['Chest Pain Type'].cat.rename_categories({1: 'Typical Angina', 2: 'Atypical Angina', 3: 'Non-Anginal Pain', 4: 'Asymptomatic'})
data_with_target['Fasting Blood Sugar'] = data_with_target['Fasting Blood Sugar'].cat.rename_categories({1: '>120mg/ml', 0: '<120mg/ml'})   
data_with_target['Resting Electrocardiographic Results'] = data_with_target['Resting Electrocardiographic Results'].cat.rename_categories({0: 'Normal', 1: 'ST-T wave abnormality', 2: 'Left Ventricular Hypertrophy'})
data_with_target['Presence of Exercise Induced Angina'] = data_with_target['Presence of Exercise Induced Angina'].cat.rename_categories({1: 'Yes', 0: 'No'})
data_with_target['ST/HR Slope of the Peak Exercise ST Segment'] = data_with_target['ST/HR Slope of the Peak Exercise ST Segment'].cat.rename_categories({1: 'Upsloping', 2: 'Flat', 3: 'Downsloping'})
data_with_target['Number of Major Vessels Colored by Flourosopy'] = data_with_target['Number of Major Vessels Colored by Flourosopy'].cat.rename_categories({0: '0', 1: '1', 2: '2', 3: '3', 4: '4'})
data_with_target['Thalassemia'] = data_with_target['Thalassemia'].cat.rename_categories({3: 'Normal', 6: 'Fixed Defect', 7: 'Reversible Defect'})

# Combine categories 1, 2, 3, and 4 into one category for 'Heart Disease'
data_with_target['Heart Disease'] = data_with_target['Heart Disease'].map({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})

# Convert 'Heart Disease' back to categorical with the new categories
data_with_target['Heart Disease'] = data_with_target['Heart Disease'].astype('category')
data_with_target['Heart Disease'] = data_with_target['Heart Disease'].cat.rename_categories({1: 'Heart Disease', 0: 'No Heart Disease'})
```
